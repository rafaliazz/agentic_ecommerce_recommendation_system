```python
import cv2
import easyocr
import numpy as np
import torch
import matplotlib.pyplot as plt

IMAGE_PATH = r"D:\personalProjects\final_proj\src\dataset\pc\shopee_1.png"  # change this

reader = easyocr.Reader(['en'], gpu=torch.cuda.is_available())

image = cv2.imread(IMAGE_PATH)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

h, w, _ = image_rgb.shape

results = reader.readtext(image_rgb)

filtered_text = []

for (bbox, text, conf) in results:
    if conf < 0.6:
        continue

    # Bounding box coordinates
    (tl, tr, br, bl) = bbox
    x_center = (tl[0] + br[0]) / 2
    y_center = (tl[1] + br[1]) / 2

    box_height = abs(bl[1] - tl[1])

    # ---- SMART FILTERS ----
    
    # 1. Ignore left image panel
    if x_center < w * 0.45:
        continue

    # 2. Ignore top navbar
    if y_center < h * 0.15:
        continue

    # 3. Ignore tiny UI text
    if box_height < 15:
        continue

    filtered_text.append((text, box_height, conf))

print("Filtered Results:\n")

for t in filtered_text:
    print(t[0])
```

```python
import cv2
import easyocr
import numpy as np
import torch

IMAGE_PATH = r"D:\personalProjects\final_proj\src\dataset\mobile\shopee_1.png"  # change this

reader = easyocr.Reader(['en'], gpu=torch.cuda.is_available())

image = cv2.imread(IMAGE_PATH)

if image is None:
    raise ValueError("Image not found.")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

h, w, _ = image_rgb.shape

results = reader.readtext(image_rgb)

candidates = []

for (bbox, text, conf) in results:
    if conf < 0.6:
        continue

    (tl, tr, br, bl) = bbox
    x_center = (tl[0] + br[0]) / 2
    y_center = (tl[1] + br[1]) / 2
    box_height = abs(bl[1] - tl[1])

    # -------- MOBILE FILTERS --------

    # 1. Ignore very top status/store bar
    if y_center < h * 0.18:
        continue

    # 2. Ignore extreme bottom nav buttons
    if y_center > h * 0.92:
        continue

    # 3. Ignore tiny UI text
    if box_height < 18:
        continue

    candidates.append({
        "text": text,
        "y": y_center,
        "height": box_height,
        "conf": conf
    })

# Sort by vertical position
candidates = sorted(candidates, key=lambda x: x["y"])

print("---- FILTERED TEXT ----\n")

for c in candidates:
    print(c["text"])
```

```python
added_text = ""
for c in candidates:
    added_text += c["text"] +"\n"
 
```
